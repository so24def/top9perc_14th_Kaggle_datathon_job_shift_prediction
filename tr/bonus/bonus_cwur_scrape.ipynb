```python
# https://cwur.org/ sitesinden üniversite sıralamalarını scrape etme

import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://cwur.org/2018-19.php"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the table you want to scrape
    table = soup.find("table")
    
    # Extract the rows from the table
    rows = table.find_all("tr")
    
    # Create a list to store the data
    data = []
    
    # Iterate through the rows and extract the data
    for row in rows:
        cells = row.find_all("td")
        data.append([cell.text for cell in cells])
    
    # Create a pandas dataframe from the data
    df = pd.DataFrame(data)
    
    # Save the data to a CSV file
    df.to_csv("cwur_table2018.csv", index=False, header=False)
else:
    print("Failed to retrieve data from the website")

    
###

url = "https://cwur.org/2017.php"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the table you want to scrape
    table = soup.find("table")
    
    # Extract the rows from the table
    rows = table.find_all("tr")
    
    # Create a list to store the data
    data = []
    
    # Iterate through the rows and extract the data
    for row in rows:
        cells = row.find_all("td")
        data.append([cell.text for cell in cells])
    
    # Create a pandas dataframe from the data
    df = pd.DataFrame(data)
    
    # Save the data to a CSV file
    df.to_csv("cwur_table2017.csv", index=False, header=False)
else:
    print("Failed to retrieve data from the website")

###    

url = "https://cwur.org/2016.php"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the table you want to scrape
    table = soup.find("table")
    
    # Extract the rows from the table
    rows = table.find_all("tr")
    
    # Create a list to store the data
    data = []
    
    # Iterate through the rows and extract the data
    for row in rows:
        cells = row.find_all("td")
        data.append([cell.text for cell in cells])
    
    # Create a pandas dataframe from the data
    df = pd.DataFrame(data)
    
    # Save the data to a CSV file
    df.to_csv("cwur_table2016.csv", index=False, header=False)
else:
    print("Failed to retrieve data from the website")
```
